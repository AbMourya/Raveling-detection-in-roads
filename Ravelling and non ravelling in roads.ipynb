```python
import os
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
import cv2
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
os.chdir("C:\\Users\\abmou\\Dropbox\\PC\\Downloads\\raveling-detection-ce784a-2023\\mod_ravelling_dataset")
ravel_train=os.listdir('train/Raveling')
non_ravel_train=os.listdir('train/Non_Raveling')
```

```python
# FOE TRAINING DATA CALCULATION
train_images = []
train_labels = []

# RAVELING DATA
for img in ravel_train:
    image = cv2.imread('train/Raveling/' + img)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    #STATISTICAL PROPERTIES
    # HISTOGRAM
    hist = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    # MEAN
    mean = np.mean(image, axis=(0, 1))
    # STANDARD DEVIATION
    std = np.std(image, axis=(0, 1))
    # RANGE
    rng= np.max(image, axis=(0,1))-np.mean(image,axis=(0,1))
    # RANGE 2
    rng2= np.max(image, axis=(0,1))-np.min(image,axis=(0,1))
    # SKEWNESS
    skewness = skew(image, axis=(0,1))
    # KURTOSIS
    kurt = kurtosis(image, axis=(0,1))
    
        # Normalize the RGB channels
    norm_image = (image - np.min(image)) / (np.max(image) - np.min(image))
    
    # Compute the mean, standard deviation, range, skewness, and kurtosis of the normalized RGB channels
    norm_mean = np.mean(norm_image, axis=(0, 1))
    norm_std = np.std(norm_image, axis=(0, 1))
    norm_rng = np.max(norm_image, axis=(0, 1)) - np.mean(norm_image, axis=(0, 1))
    norm_rng2 = np.max(norm_image, axis=(0, 1)) - np.min(norm_image, axis=(0, 1))
    norm_skewness = skew(norm_image, axis=(0, 1))
    norm_kurt = kurtosis(norm_image, axis=(0, 1))
    
    # Concatenate the features into a single feature vector
    features = np.concatenate([mean, std, rng, rng2, skewness, kurt, norm_mean, norm_std, norm_rng, norm_rng2, norm_skewness, norm_kurt, hist.ravel()]) 
    train_images.append(features)
    train_labels.append(1)
    
 # NON RAVELING DATA   
for img in non_ravel_train:
    image = cv2.imread('train/Non_Raveling/' + img)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    #STATISTICAL PROPERTIES
    # HISTOGRAM
    hist = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    # MEAN
    mean = np.mean(image, axis=(0, 1))
    # STANDARD DEVIATION
    std = np.std(image, axis=(0, 1))
    # RANGE
    rng= np.max(image, axis=(0,1))-np.mean(image,axis=(0,1))
    # RANGE 2
    rng2= np.max(image, axis=(0,1))-np.min(image,axis=(0,1))
    # SKEWNESS
    skewness = skew(image, axis=(0,1))
    # KURTOSIS
    kurt = kurtosis(image, axis=(0,1))
    
    
        # Normalize the RGB channels
    norm_image = (image - np.min(image)) / (np.max(image) - np.min(image))
    
    # Compute the mean, standard deviation, range, skewness, and kurtosis of the normalized RGB channels
    norm_mean = np.mean(norm_image, axis=(0, 1))
    norm_std = np.std(norm_image, axis=(0, 1))
    norm_rng = np.max(norm_image, axis=(0, 1)) - np.mean(norm_image, axis=(0, 1))
    norm_rng2 = np.max(norm_image, axis=(0, 1)) - np.min(norm_image, axis=(0, 1))
    norm_skewness = skew(norm_image, axis=(0, 1))
    norm_kurt = kurtosis(norm_image, axis=(0, 1))
    skewness = skew(image, axis=(0,1))
    kurt = kurtosis(image, axis=(0,1))
    
    # Concatenate the features into a single feature vector
    features = np.concatenate([mean, std, rng, rng2, skewness, kurt, norm_mean, norm_std, norm_rng, norm_rng2, norm_skewness, norm_kurt, hist.ravel()])

    train_images.append(features)
    train_labels.append(0) 
train_images = np.array(train_images)
train_labels = np.array(train_labels) 
```

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Train the Logistic Regression model
model = LogisticRegression(max_iter =5000)
model.fit(train_images, train_labels)
test_images=[]
filenames=[]
for i in range(1, 301):
    image = cv2.imread('test/' + str(i) + '.jpg')
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
     #STATISTICAL PROPERTIES
    # HISTOGRAM
    hist = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    # MEAN
    mean = np.mean(image, axis=(0, 1))
    # STANDARD DEVIATION
    std = np.std(image, axis=(0, 1))
    # RANGE
    rng= np.max(image, axis=(0,1))-np.mean(image,axis=(0,1))
    # RANGE 2
    rng2= np.max(image, axis=(0,1))-np.min(image,axis=(0,1))
    # SKEWNESS
    skewness = skew(image, axis=(0,1))
    # KURTOSIS
    kurt = kurtosis(image, axis=(0,1))
    
    
        # Normalize the RGB channels
    norm_image = (image - np.min(image)) / (np.max(image) - np.min(image))
    
    # Compute the mean, standard deviation, range, skewness, and kurtosis of the normalized RGB channels
    norm_mean = np.mean(norm_image, axis=(0, 1))
    norm_std = np.std(norm_image, axis=(0, 1))
    norm_rng = np.max(norm_image, axis=(0, 1)) - np.mean(norm_image, axis=(0, 1))
    norm_rng2 = np.max(norm_image, axis=(0, 1)) - np.min(norm_image, axis=(0, 1))
    norm_skewness = skew(norm_image, axis=(0, 1))
    norm_kurt = kurtosis(norm_image, axis=(0, 1))
    
    # Concatenate the features into a single feature vector
    features = np.concatenate([mean, std, rng, rng2, skewness, kurt, norm_mean, norm_std, norm_rng, norm_rng2, norm_skewness, norm_kurt,hist.ravel()])

       
    test_images.append(features)
    filenames.append(str(i) + '.jpg')
test_labels2=[]
test_images = np.array(test_images)

test_labels = model.predict(test_images)
```
